## Data Visualization:-

 ### 1. Distribution of Cybercrime Incident Types

```python
plt.figure(figsize=(8,5))
df['Incident_Type'].value_counts().plot(kind='bar')

plt.title("Distribution of Cybercrime Incident Types")
plt.xlabel("Incident Type")
plt.ylabel("Number of Cases")

plt.show()
```

####  * This chart identifies the most common cybercrime types.
####  * Certain attack methods occur significantly more frequently than others.
####  * Frequent attack types indicate major cybersecurity threats.
####  * Organizations should focus prevention efforts on these attack categories.

### 2. Cybercrime cases across different categories

```python
plt.figure(figsize=(8,5))
df['Category'].value_counts().plot(kind='bar')

plt.title("Cybercrime Cases Across Different Categories")
plt.xlabel("Category")
plt.ylabel("Number of Cases")

plt.show()
```

#### - Some sectors are targeted more frequently than others.
#### - The most affected category requires stronger cybersecurity measures.
#### - Attackers often focus on sectors with valuable information.
#### - This helps identify vulnerable industries.

### 3.  Year-wise financial loss due to cybercrime

```python
plt.figure(figsize=(8,5))
df.groupby('Year')['Amount_Lost_INR'].sum().plot(kind='line', marker='o')

plt.title("Year-wise Financial Loss Due to Cybercrime")
plt.xlabel("Year")
plt.ylabel("Total Amount Lost (INR)")

plt.show()
```

- #### Financial losses change noticeably across years.
- #### Increasing trends indicate growing cybercrime impact.
- #### Decreasing trends may reflect improved security awareness.
- #### Trend analysis helps predict future risks.

### 4. City wise total Financial Loss by Cybercrime

```python
plt.figure(figsize=(10,5))
df.groupby('City')['Amount_Lost_INR'].sum().plot(kind='bar')

plt.title("City-wise Total Financial Loss from Cybercrime")
plt.xlabel("City")
plt.ylabel("Total Amount Lost (INR)")

plt.show()
```

- #### Some cities contribute heavily to overall losses.
- #### Urban areas often experience more cybercrime incidents.
- #### High-loss cities may require stronger awareness campaigns.
- #### The chart highlights cybercrime hotspots.

### 5. Percentage share of Cybercrime Incident Types

```python
plt.figure(figsize=(8,8))
df['Incident_Type'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Percentage Share of Cybercrime Incident Types")
plt.ylabel("")

plt.show()
```

- ####  This chart shows percentage contribution of each cybercrime type.
- #### A few attack types dominate total incidents.
- #### Helps understand the composition of cyber threats.
- #### Useful for prioritizing security strategies.

### 6. Sector-wise Distribution of Cybercrime Cases

```python
plt.figure(figsize=(8,8))
df['Category'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Sector-wise Distribution of Cybercrime Cases")
plt.ylabel("")

plt.show()
```

- #### Displays how incidents are distributed across sectors.
- #### Some categories occupy a major share of attacks.
- #### Indicates where cybersecurity investments are most needed.
- #### Helps identify vulnerable sectors.

### 7. Relationship between Incident Day and Financial Loss

```python
plt.figure(figsize=(8,5))
plt.scatter(
    df['Day'],
    df['Amount_Lost_INR']
)

plt.title("Relationship Between Incident Day and Financial Loss")
plt.xlabel("Day")
plt.ylabel("Amount Lost (INR)")

plt.show()
```

- #### Shows relationship between incident day and financial loss.
- #### Data points are spread across different values.
- #### No strong pattern may indicate random occurrence of losses.
- #### Outliers represent exceptionally costly incidents.

### 8. Year wise number of Cybercrime Incidents

```python
plt.figure(figsize=(8,5))

df.groupby('Year').size().plot(
    kind='line',
    marker='o'
)

plt.title("Year-wise Number of Cybercrime Incidents")
plt.xlabel("Year")
plt.ylabel("Number of Incidents")

plt.show()
```

- #### This chart shows how the number of cybercrime incidents has changed over the years.
- #### An increasing trend indicates growing cybercrime activity and reporting.
- #### Peaks in certain years may be associated with increased digital adoption or emerging threats.
- #### The analysis helps understand long-term cybercrime trends.

### 9. Outlier Analysis of Financial Losses

```python
plt.figure(figsize=(8,5))
sns.boxplot(
    y=df['Amount_Lost_INR']
)

plt.title("Outlier Analysis of Financial Losses")
plt.ylabel("Amount Lost (INR)")

plt.show()
```

- #### This chart clearly identifies outliers in financial losses.
- #### Most incidents are concentrated within a narrow range.
- #### A few cases result in exceptionally high losses.
- #### These outliers significantly affect overall averages.

### 10. Correlation Analysis of Numerical Variables

```python
plt.figure(figsize=(8,5))
sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Analysis of Numerical Variables")

plt.show()
```

- ####  This chart Shows relationships between numerical variables.
- #### Strong correlations indicate connected behavior.
- #### Weak correlations suggest independent variables.
- #### Useful for feature analysis and understanding data structure.
